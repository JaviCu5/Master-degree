```python
import matplotlib.pyplot as plt
import scipy as s
import numpy as np
import statistics as st

fichero = open('Weibull.dat', 'r')
data = []
for i in fichero:
    data.append(float(i.split()[0]))
    

def gamma(x) -> float:
    return s.special.gamma(x)

def f(alpha):
    return (gamma(1+2/alpha)/gamma(1+1/alpha)**2) + coef
    
media = np.array(data).mean()
var = np.array(data).var()
coef = -1 - var/media**2
proposedAlpha = 1.5

#Forma alpha
#Escala lambda
Alpha = s.optimize.newton(f, proposedAlpha)
Lambda = media/gamma(1+1/Alpha)


#Analytic behaviour
x = np.linspace(0., 5., 501)
def y(X, escala, forma) -> [float]:
    return (escala*forma)*(escala*x)**(forma - 1.)*np.exp(-(x*escala)**(forma))

fig, ax = plt.subplots()

fig.set_size_inches(10, 5)

ax.set_title('Weibull distribution')
ax.set_xlabel('x')
ax.set_ylabel('ρ(x)')
ax.plot(x, y(x, Lambda, Alpha))
ax.hist(data, density = True, bins=200, color='#F2AB6D', rwidth=0.85)

ax.legend(['α = 1.5668 λ = 0.997','Raw data'], frameon=False, loc=1)

plt.xlim(0, 5)
plt.savefig("WeibullCaracterized1.jpg")
plt.show()
```
